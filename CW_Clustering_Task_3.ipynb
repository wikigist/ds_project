```python
# Performing Normalization with two chosen data from our dataset

from sklearn.preprocessing import StandardScaler
#X = pokemon_data.iloc[:, [8,10]].values
#sc_X = StandardScaler()
#X = sc_X.fit_transform(X)


sc_X = StandardScaler()
X = sc_X.fit_transform(pokemon_data[Skills])
```

## K-Means Clustering

```python
# First, using the Elbow method to check for the optimal number of clusters in our data via K-Means Clustering


from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    #pokemon_data["Clusters"] = kmeans.labels_
    
plt.plot(range(1, 11), wcss, linewidth=2, marker='8')
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Checking Silhouette coefficient for more clarification

from sklearn.metrics import silhouette_score

label = kmeans.predict(X)
#kmeans.fit(X)

# Calculating Silhouette score

print(f' Silhouette Score(n=4): {silhouette_score(X, label)}')
```

```python
# Now that we've identified the number of clusters to be 4, we can now fit the KMeans to the dataset

kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Now we visualise our clusters using scatterplot

from pylab import *
rcParams['figure.figsize'] = 16,10

#plt.figure(figsize(8,8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'First Cluster')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Second Cluster')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Third Cluster')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'black', label = 'Fourth Cluster')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids' )


plt.title("Clusters of Pokemon Agility")
plt.xlabel("Special Attack(Scaled)")
plt.ylabel("Speed(Scaled)")
plt.legend()
plt.show()
```

## Hierarchical Clustering

```python
# Checking to see if our data is still in place

pokemon_data.head()
```

```python
"""
Now using the Scipy package to visualise the dendrogram of our data 
hence bringing out the optimal number of clusters

""" 
rcParams['figure.figsize'] = 15,6
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title("Dendogram")
plt.xlabel("Pokemon Agility")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
# Now we fit hierarchical clustering into dataset using "AgglomerativeClustering"

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```
