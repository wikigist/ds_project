#### Evaluating the model

```python
# Evaluating the model by using the predict function to make predictions on our test data

y_pred = knn_classifier.predict(X_test)
print(y_pred)
```

```python
X_test.head(1)
```

```python
# Evaluating model using confusion matrix

from sklearn import metrics 


acc = metrics.accuracy_score(y_test,y_pred)
print('accuracy:%.2f\n\n'%(acc))
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm, "\n\n")
print("-----------------------------------------------------")
result = metrics.classification_report(y_test, y_pred)
print("Classification Report:\n",)
print(result)
```

```python
# Visualizing confusion matrix using seaborn heatmap for more insight

#ax = sns.heatmap(cm, cmap='flare', annot=True, fmt='d')

vmin = np.min(cm)
vmax = np.max(cm)
off_diag_mask = np.eye(*cm.shape, dtype=bool)

ax = sns.heatmap(cm, annot=True, mask=~off_diag_mask, cmap='Blues', vmin=vmin, vmax=vmax)
ax = sns.heatmap(cm, annot=True, mask=off_diag_mask, cmap='OrRd', vmin=vmin, vmax=vmax, cbar_kws=dict(ticks=[]))

#plt.figure(figsize=())
plt.xlabel("Predicted Class", fontsize=15)
plt.ylabel("True Class", fontsize=15)
plt.title("Confusion Matrix", fontsize=15)

plt.show()
```

```python
# Applying cross-validation for parameter tuning on Knn

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

#creating list of K for KNN

k_list = list(range(1,50,2))

# creating list of cv scores

cv_scores = []

# perform 10-fold cross validation

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
# changing to misclassification error

miss_classify = [1 - x for x in cv_scores]

plt.figure()
plt.figure(figsize=(9,6))
plt.title('Number of neighbors', fontsize=15, fontweight='bold')
plt.xlabel('Number of Neighbors K', fontsize=12)
plt.ylabel('Misclassification Error', fontsize=12)
sns.set_style("whitegrid")
plt.plot(k_list, miss_classify)

plt.show()
```

```python
# Now, we find best k

optimal_k = k_list[miss_classify.index(min(miss_classify))]
print("Number of neighbors is %d!" % optimal_k)
```

## Instantiating & Training Model

```python
# Now we train our model using Keras tensorflow

import numpy as np
from keras.utils import np_utils

Y_train = np_utils.to_categorical(Y_train)
print(Y_train.shape)
Y_test_c = np_utils.to_categorical(Y_test)
```

```python
# Instantiating our model

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy

nn_model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=5, input_shape=(4,), activation='relu'),
    tf.keras.layers.Dense(units=25, activation='sigmoid'),
    tf.keras.layers.Dense(units=25, activation='sigmoid'),
    tf.keras.layers.Dense(units=25, activation='sigmoid'),
    tf.keras.layers.Dense(units=3, activation='sigmoid')
])

nn_model.compile(loss="categorical_crossentropy", optimizer='adam', metrics='accuracy')
```

```python
nn_model.summary()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# training weighted classes to identify observations

#history = nn_model.fit(X_train, Y_train, validation_data =(X_test, Y_test), epochs=50)

#history = nn_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=0)
```

```python
class_weights = {0:1, 1:6, 2:7}

#class_weights = {0:1,1:2.}
history = nn_model.fit(X_train, Y_train, \
                       batch_size = 32, epochs= 400, \
                       verbose=2, class_weight=class_weights, \
                       validation_split=0.2)
```

```python
(X_test, Y_test_c)
```

## Evaluating Neural Network

```python
"""
Reviewing the accuracy and validation metrics of our neural network & 
also plotting on the training and validation accuracy

"""

accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']


plt.plot(accuracy, label='Training Set Accuracy')
plt.plot(validation_accuracy, label='Validation Set Accuracy')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy Across Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# checking for the loss

loss = history.history['loss']
validation_loss = history.history['val_loss']

plt.plot(loss, label='Training Set Loss')
plt.plot(validation_loss, label='Validation Set Loss')
plt.ylabel('Loss')
plt.title('Training and Validation Accuracy Across Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
